```python
from ewatercycle.observation.grdc import get_grdc_data
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from cartopy import crs
from cartopy import feature as cfeature
import matplotlib.colors as mcolors
```

```python


grdc_chatly = get_grdc_data(2817100,
                   '1900-01-01T00:00Z',
                   '2001-01-01T00:00Z',
                   data_home='/home/avandervee3/MSc_AralSea/book/thesis_projects/MSc/2025_Q1_AndreVanDerVeen_CEG/work_in_progress/Test_Aral/')

grdc_kerki = get_grdc_data(2617110,
                   '1900-01-01T00:00Z',
                   '2001-01-01T00:00Z',
                   data_home='/home/avandervee3/MSc_AralSea/book/thesis_projects/MSc/2025_Q1_AndreVanDerVeen_CEG/work_in_progress/Test_Aral/')
```

```python
plt.figure(figsize=(10,5))
grdc_chatly['streamflow'].plot(label='Chatly')
grdc_kerki['streamflow'].plot(label='Kerki')
plt.title('GRDC Discharge Data for Chatly and Kerki Stations')

```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4), sharey=False)

# --- LEFT: Full time series ---
grdc_chatly['streamflow'].plot(ax=ax[0], label='Chatly (downstream)', color='C0')
grdc_kerki['streamflow'].plot(ax=ax[0], label='Kerki (upstream)', color='C1')
ax[0].set_title('Full Streamflow Record')
ax[0].set_xlabel('Date')
#ax[0].set_ylabel('Streamflow [m/s]')
ax[0].legend()
ax[0].grid(True, linestyle='--', alpha=0.5)

# --- RIGHT: Only 1960 ---
grdc_chatly.sel(time=slice('1960-01-01', '1960-12-31'))['streamflow'].plot(
    ax=ax[1], label='Chatly (downstream)', color='C0'
)
grdc_kerki.sel(time=slice('1960-01-01', '1960-12-31'))['streamflow'].plot(
    ax=ax[1], label='Kerki (upstream)', color='C1'
)
ax[1].set_title('Streamflow in 1960')
ax[1].set_xlabel('Date')
ax[1].legend()
ax[1].grid(True, linestyle='--', alpha=0.5)

plt.suptitle('Daily Average Streamflow at GRDC station Kerki (upstream) and GRDC station Chatly (downstream)', fontsize=12)
plt.tight_layout()
plt.savefig('GRDC_Streamflow_Chatly_Kerki.png', bbox_inches='tight', pad_inches=0.05, dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Plot GRDC streamflow for the year 1960
grdc_chatly.loc['1960', 'streamflow'].plot(label='Chatly (GRDC)')
grdc_kerki.loc['1960', 'streamflow'].plot(label='Kerki (GRDC)')

plt.title('Daily Streamflow in 1960 at Chatly and Kerki (GRDC Stations)')
plt.xlabel('Date')
plt.ylabel('Streamflow [m/s]')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
